```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("genders.csv", index_col=0)
df.head()
denman_df = df["Denman"]
hornby_df = df["Hornby"]
cv_df = df["CV"]
bc_df = df["BC"]
canada_df = df["Canada"]
```

```python
denman_df.plot(kind="pie", autopct="%1.2f%%",
               colors=sns.color_palette("Blues", 2), ylabel="")
```

```python
cv_df.plot(kind="pie", autopct="%1.2f%%",
           colors=sns.color_palette("Reds", 2), ylabel="")
```

```python
hornby_df.plot(kind="pie", autopct="%1.2f%%",
               colors=sns.color_palette("Purples", 2), ylabel="")
```

```python
bc_df.plot(kind="pie", autopct="%1.2f%%",
           colors=sns.color_palette("Oranges", 2), ylabel="")
```
